```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("MBA Salary.csv")
df.head(4)
```

```python
# conver it into a list

Xactual=df["Percentage in Grade 10"].tolist()
Yactual=df["Salary"].tolist()
```

```python
Yaverage=np.mean(Yactual) ## ymean
Xaverage=np.mean(Xactual) ##xmean

print(Xaverage,Yaverage)
```

```python
nr=[]
dr=[]
for inx in range(len(Xactual)):
    Xin=Xactual[inx]
    Yin=Yactual[inx]
    Xdiv=Xin-Xaverage
    Ydiv=Yin-Yaverage
    nr.append(Ydiv*Xdiv)
    dr.append(Xdiv**2)
    #print(nr,dr)
slopem=sum(nr)/sum(dr)
print(slopem)
```

```python
## intercept
c=Yaverage-slopem*Xaverage
print(c)
```

```python
ypred=[]
for i in range(len(Xactual)):
    Ypred=(slopem*Xactual[i]+c)
    #print(Ypred)  
    ypred.append(Ypred)
            
```

```python
df["Ypredicted"]=ypred
df.head(3)
```

```python
# Calculate r2

#1-nr(Yactual-Ypredicted)/Y(actual)-Yaverage
```

```python
nre=[]
dre=[]
for i in range(len(Xactual)):
    nr=(Yactual[i]-ypred[i])
    nre.append(nr**2)
    dr=(Yactual[i]-Yaverage)
    dre.append(dr**2)
r2=sum(nre)/sum(dre)
print("r2 error is...:",1-r2)
```
