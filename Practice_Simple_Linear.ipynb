```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("headbrain.csv")
df.tail(4)
```

```python
xa=df["Head Size(cm^3)"].tolist()
ya=df["Brain Weight(grams)"].tolist()
```

```python
xmean=np.mean(xa)
ymean=np.mean(ya)
print(xmean,ymean)
```

```python
den=[]
xya=[]
for inx in range(len(xa)):
    #print(inx,xa[inx])
    xi=xa[inx]
    yi=ya[inx]
    #print(inx,xi)
    #print(inx,yi)
    disy=yi-ymean
    #print(disy)
    disx=xi-xmean
    #print(disx)
    x2=(disx**2)
    #print(x2)
    
    xya.append(disx*disy)
    den.append(x2)
slopem=sum(xya)/sum(den)
print(slopem)

C=ymean-slopem*xmean
print(C)
    
   
```

```python
Ypredict=[]
for i in range(len(xa)):
    ypredict=slopem*xa[i]+C
    Ypredict.append(ypredict)
    print(ypredict)
df["Ypredict"]=Ypredict
```

```python
df["Yactual"]=ya
```
