```python
df = pd.read_csv('headbrain.csv')
```

```python
X = df['Head Size(cm^3)'].tolist()
Y = df['Brain Weight(grams)'].tolist()
```

```python
print(len(X),len(Y))
```

```python
xAvg = np.mean(X) #
print('X avg is ...',xAvg)
```

```python
yAvg = np.mean(Y) #
print('Y avg is ...',yAvg)
```

```python
# YYa = []
# XXa = []
XYa = []
Den =[]
for inx in range(len(X)):
    xI = X[inx]
    yI = Y[inx]
#     print(inx,xI,yI)
    distXI = xI-xAvg
    distYI = yI-yAvg
#     print(distXI,distYI)
    x2 = distXI**2
#     print(x2)
    XYa.append(distXI*distYI)
    Den.append(x2)

slopeM = sum(XYa)/sum(Den)
print(slopeM)

C = yAvg - slopeM*xAvg
print('C is...',C)
    
```

```python
listYPred = []
for inxP in range(len(X)):
    yPred = slopeM*X[inxP]+C
    print(Y[inxP],yPred)
    listYPred.append(yPred)
```

```python
df['Brain Weight Predicted'] = listYPred
```
